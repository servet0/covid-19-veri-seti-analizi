```python
columns_needed = ['Accident_Index', 'Number_of_Vehicles', 'Number_of_Casualties', 'Accident_Severity', 'Date', 'Weather_Conditions', 'Road_Surface_Conditions', 'Urban_or_Rural_Area']
```

```python
df = df[columns_needed]
```

```python
df.head()
```

```python
df.tail()
```

```python
df['trafic_flow_category'] = pd.cut(df['Number_of_Vehicles'], bins=[1,2,3, df['Number_of_Vehicles'].max()], labels=['Low', 'Medium', 'High'])
```

```python
accidents_by_traffic = df.groupby('trafic_flow_category')['Accident_Index'].count()
```

```python
accidents_by_traffic
```

```python
accidents_by_traffic.plot()
plt.xlabel('Trafik Akış Kategorisi')
plt.ylabel('Kaza Sayısı')
plt.title('Trafik Akışşına Göre Kaza Sayısı')
```

```python
accidents_by_traffic.plot(kind='bar')
plt.xlabel('Trafik Akış Kategorisi')
plt.ylabel('Kaza Sayısı')
plt.title('Trafik Akışşına Göre Kaza Sayısı')
```

```python
accidents_by_traffic.plot(kind='pie')
plt.xlabel('Trafik Akış Kategorisi')
plt.ylabel('Kaza Sayısı')
plt.title('Trafik Akışşına Göre Kaza Sayısı')
```
